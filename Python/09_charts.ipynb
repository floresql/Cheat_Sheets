## âœ¨ Why This Matters
Different chart types serve different purposes. Choosing the right one helps you tell a clearer story, avoid misleading conclusions, and impress stakeholders who pretend they understand statistics.

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
````

## ðŸ“Š Bar Chart â€” Comparison
Use bar charts to compare categories. Best when you have a few categories and want to highlight magnitude differences.

**When to use:** Comparing sales across regions, response counts, etc.

````python
def bar_chart():
    categories = ['A', 'B', 'C']
    values = [23, 45, 56]
    plt.bar(categories, values, color='skyblue')
    plt.title('Bar Chart Example')
    plt.xlabel('Category')
    plt.ylabel('Value')
    plt.show()

bar_chart()
````

## ðŸ“ˆ Histogram â€” Distribution
Histograms are perfect for understanding the spread and frequency of your data values.

**When to use:** Examining salary distribution, scores, or any continuous numeric variable.

````python
def histogram():
    data = np.random.normal(0, 1, 1000)
    plt.hist(data, bins=30, color='salmon')
    plt.title('Histogram Example')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

histogram()
````

## ðŸ¥§ Pie Chart â€” Composition
Pie charts help show how parts make up a whole. Use sparingly and only with a few categories.

**When to use:** Market share, category splits, or budget allocation.

````python
def pie_chart():
    labels = ['Product A', 'Product B', 'Product C']
    sizes = [40, 35, 25]
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title('Pie Chart Example')
    plt.axis('equal')
    plt.show()

pie_chart()
````

## ðŸ”„ Scatter Plot â€” Relationship
Scatter plots show relationships or correlations between two variables. Theyâ€™re essential for spotting trends, clusters, or outliers.

**When to use:** Salary vs experience, height vs weight, or any X vs Y combo.

````python
def scatter_plot():
    x = np.random.rand(50)
    y = 2 * x + np.random.normal(0, 0.1, 50)
    plt.scatter(x, y, color='lime')
    plt.title('Scatter Plot Example')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()

scatter_plot()
````

## ðŸ•’ Line Chart â€” Change Over Time
Line charts are your go-to for showing data over time. Great for tracking metrics, trends, or forecasting.

**When to use:** Stock prices, monthly sales, user growth.

````python
def line_chart():
    time = np.arange(1, 11)
    value = np.random.randint(10, 100, size=10)
    plt.plot(time, value, marker='o')
    plt.title('Line Chart Example')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.grid(True)
    plt.show()

line_chart()
````
